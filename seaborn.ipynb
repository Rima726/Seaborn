```python
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#distribution graphs
#----------->1.Displot
#----------->2.jointplot
#----------->3.pairplot
#----------->4.rugplot

#categorical plots
#----------->1,barplot
#----------->2.countplot
#----------->3.boxplot
#----------->viol inplot


```

# distplot

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

# pair plot

```python
sns.pairplot(tips,hue='sex')
```

# rag plot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.countplot(tips['smoker'],data=tips,hue='sex')
```

```python
import numpy as np
sns.barplot(x='day',y='tips',data=tips,hue='sex',estimator=np.median)
```

```python
sns.boxplot('total_bill',data=tips)
```

```python
sns.boxplot('total_bill',data=tips,hue='smoker')
```

```python
sns.boxplot(x=tips['sex'],y=tips['tip'])
```

```python
sns.violinplot(x=tips['sex'],y=tips['tip'],split=False)
```
