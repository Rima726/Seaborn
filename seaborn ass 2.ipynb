```python

iris = sns.load_dataset('iris')
```

```python
iris.head(5)
```

```python
sns.pairplot(iris,hue='species')
```

In the above each and every column or feature is plotted with each and every column(scatter plot) and same column is also plotted(histogram) in x and y. Our aim is to differentiate between the three species of iris flowers and we will conclude which features can differntiate the species from the above graph.

1.Histograms: There are four histogram plotted in the above from which histogram of Sepal Width and Sepal Lenght has a big overlapping region so they will give more errors to differentiate the species but the Petal Length and Petal width has a good clear-cut plotting so they will give less errors to differentiate the species and from that Petal Lenght is the best as the setosa species is not overlapping with versicolor so we can detect the setosa flower with 100% accuracy but in Petal width the the setosa is overlapping slightly with versicolor which can make errors.

2.Scatter Plot: There are 12 scatter plots from which two scatter plots are having a linear relationship they are Petal Length vs Petal Width and another is Petal Width vs Petal length and they both are good to analysis but in the Petal Length vs Petal Width has point more close to best fit line and therefore the Petal Length vs Petal Width is the best scatter plot to differentiate the species.

So we can conclude Petal Length and Petal Width are the best feature to differentiate the flower speices.

# Problem 4:
Plot a pie chart of survived based on sex. titanic=sns.load_dataset('titanic')

```python
titanic=sns.load_dataset('titanic')
```

```python
mask1=titanic['survived']==1
d=titanic[mask1]
new=d.groupby('sex')['survived'].count()
```

```python

new.plot(kind='pie',autopct='%0.2f')
```
